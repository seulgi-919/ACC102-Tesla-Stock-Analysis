# Downloading Tesla (TSLA) historical stock data for the past 5 years

```python
ticker = "TSLA"
```

```python
tesla = yf.Ticker(ticker)
```

```python
# Get historical market data
```

```python
df_stock = tesla.history(period="5y")
```

```python
# Get quarterly financial statements
```

```python
df_financials = tesla.quarterly_financials.T 
```

```python
print("Data successfully downloaded from Yahoo Finance.")[cite: 1]
```

# Calculating 50-day and 200-day Moving Averages (MA) for trend analysis

```python
df_stock['MA50'] = df_stock['Close'].rolling(window=50).mean()
```

```python
df_stock['MA200'] = df_stock['Close'].rolling(window=200).mean()
```

```python
# Cleaning financial data: selecting Revenue and Net Income
```

```python
# Note: Financial data is often in different units; we convert to Billions for clarity.
```

```python
business_perf = df_financials[['Total Revenue', 'Net Income']].copy()
```

```python
business_perf.index = pd.to_datetime(business_perf.index)
```

```python
business_perf = business_perf.sort_index() / 1e9 # Convert to Billions USD
```

```python
print("Data transformation complete.")[cite: 1]
```

# Plotting the Stock Price with Moving Averages

```python
plt.figure(figsize=(14, 7))
```

```python
plt.plot(df_stock.index, df_stock['Close'], label='TSLA Close Price', alpha=0.6, color='blue')
```

```python
plt.plot(df_stock.index, df_stock['MA50'], label='50-day Moving Average', color='orange', linestyle='--')
```

```python
plt.plot(df_stock.index, df_stock['MA200'], label='200-day Moving Average', color='red', linewidth=2)
```

```python
plt.title(f'{ticker} Stock Price Trend & Momentum (Last 5 Years)', fontsize=15)
```

```python
plt.xlabel('Date')
```

```python
plt.ylabel('Price (USD)')
```

```python
plt.legend()
```

```python
plt.grid(True, alpha=0.3)
```

```python
plt.show()
```

### ### Key Insights & Business Analysis

**1. Price Momentum:** 
The gap between the 50-day and 200-day Moving Averages indicates the current market sentiment for TSLA. A 'Golden Cross' or 'Death Cross' observed in the charts provides technical signals for potential investors.


**2. Profitability & Margins:** 
By comparing the Net Income bars against Total Revenue, we can observe the impact of Tesla's pricing strategy. While revenue may grow, the narrowing of net income reflects the competitive pressure and price wars in the EV market.


**3. Data Reliability & Limitations:** 
The data is sourced from Yahoo Finance (accessed on April 24, 2026). While reliable for educational purposes, it is subject to market volatility and potential accounting adjustments in quarterly reports.
