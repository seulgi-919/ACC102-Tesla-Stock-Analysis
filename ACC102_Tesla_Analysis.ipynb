# ----------------------# 1. Problem Definition# ----------------------print("Analytical Purpose: Analyze Tesla stock price trend, volume and volatility")print("Target Users: Individual investors & finance students")

# ----------------------# 2. Load Data# ----------------------

```python
data = yf.download("TSLA", start="2022-01-01", end="2026-04-20")
```

```python
print("Data downloaded successfully")
```

```python
data.head()
```

# ----------------------# 3. Data Cleaning (Pandas)# ----------------------

```python
data = data.dropna()
```

```python
data = data.drop_duplicates()
```

```python
data['Daily_Return'] = data['Close'].pct_change()
```

```python
print("Missing values after cleaning:\n", data.isnull().sum())
```

# ----------------------# 4. Descriptive Analysis# ----------------------print("\nDescriptive Statistics:")print(data[['Close', 'Volume', 'Daily_Return']].describe())


# ----------------------# 5. Visualization# ----------------------


```python
plt.figure(figsize=(10,4))
```

```python
plt.plot(data['Close'], label='TSLA Close Price')
```

```python
plt.title('Tesla Stock Price 2022-2026')
```

```python
plt.xlabel('Date')
```

```python
plt.ylabel('USD')
```

```python
plt.legend()
```
